```python
import cic_tokenizer

# 1. Create a dataset with some obvious structural repetition
text = "information geometry is the application of the techniques of differential geometry. " * 50

print("Initializing the topological manifold...")
tokenizer = cic_tokenizer.CICTokenizer(text)
print(f"Starting Vocabulary Size: {tokenizer.vocab_size()}")

# 2. Run the O(1) exact geometric optimization
print("\nExecuting Exact CIC phase transitions...")
merges = tokenizer.train_fast(1000)
print(f"Optimization halted naturally after {merges} merges.")
print(f"Final Vocabulary Size: {tokenizer.vocab_size()}")

# 3. Test the Encoder and Decoder
test_sentence = "exact cic approximates the topological volume."
print(f"\nOriginal text: '{test_sentence}'")

tokens = tokenizer.encode(test_sentence)
print(f"Encoded Token IDs: {tokens}")

reconstructed_text = tokenizer.decode(tokens)
print(f"Decoded text: '{reconstructed_text}'")
```
